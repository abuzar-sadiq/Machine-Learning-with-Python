```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dandrogram')
plt.xlabel('Customers')
plt.ylabel('euclidean distance')
plt.show()
```

We use dendrogram to find number of Clusters

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 5 , affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc== 0,0],X[y_hc == 0,1] , s=100, color='red',label='cluster1')
plt.scatter(X[y_hc== 1,0],X[y_hc == 1,1] , s=100, color='green',label='cluster2')
plt.scatter(X[y_hc== 2,0],X[y_hc == 2,1] , s=100, color='blue',label='cluster3')
plt.scatter(X[y_hc== 3,0],X[y_hc == 3,1] , s=100, color='cyan',label='cluster4')
plt.scatter(X[y_hc== 4,0],X[y_hc == 4,1] , s=100, color='orange',label='cluster5')
plt.title('Clusters of customs')
plt.xlabel('Annual income')
plt.ylabel('spending score')
plt.show()
```
